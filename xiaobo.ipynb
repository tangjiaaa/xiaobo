```python
import numpy as np
import pandas as pd
import os
import sys
import matplotlib.pyplot as plt

```

```python
os.getcwd()
```

```python
df = pd.read_csv('xiaobo.csv',header = None)
```

```python
l = []
for i in range(len(df.values)):
    for j in df.values[i]:
        line = j.split('\t')
        l.append(line)
```

```python
df = pd.DataFrame(np.array(l))
df.columns = df.iloc[0,:]
df.drop(0,inplace =True)
y_name = 'TotalAsset'
df
```

```python
for col in df.columns:
    if col !='Time':
        df[col] = df[col].apply(lambda x:np.nan if x=='' else float(x))
pnl = df.loc[:,['Time',y_name]]
pnl.dropna(inplace=True)
pnl.plot(x='Time',y = y_name)
plt.show()
```
